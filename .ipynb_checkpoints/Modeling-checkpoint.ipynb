### Downsample data using sklearn.utils.resample because of its simiplicity. May use penalized SVM or SMOTE for future resampling methods.

```python
#sample Fiction w/o replacement
df_majority = df[df.genre_code == 3]
df_minority = df[df.genre_code != 3]
df_majority_downsampled = resample(df_majority, replace=False, n_samples=300, random_state=240)
df_sampled = pd.concat([df_majority_downsampled, df_minority])

#sample Fantasy w/o replacement
df_majority = df_sampled[df_sampled.genre_code == 2]
df_minority = df_sampled[df_sampled.genre_code != 2]
df_majority_downsampled = resample(df_majority, replace=False, n_samples=300, random_state=240)
df_sampled = pd.concat([df_majority_downsampled, df_minority])

#sample Young Adult w/o replacement
df_majority = df_sampled[df_sampled.genre_code == 10]
df_minority = df_sampled[df_sampled.genre_code != 10]
df_majority_downsampled = resample(df_majority, replace=False, n_samples=300, random_state=240)
df_sampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
#plot the resampled distribution
plt.subplots(figsize=(15,4))
plt.xticks(rotation=0, ha='center') 
sns.countplot(x='genre', data = df_sampled, palette='Blues_r', 
              order = ['Fiction', 'Fantasy','Young Adult'
                       ,'Nonfiction','Classics','Mystery',
                       'Historical','Science Fiction','Romance',
                       'Horror','Childrens'] );
```

### Set Target Variable

```python
y = df['genre_code']
X = df['description']

#split training data into a test & train set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5,
                                                    random_state = 87302, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape )
```

```python
#since Fiction is the largest category the baseline for this model is the baseline for predicting fiction
baseline = y.value_counts()[3]/sum(y.value_counts())
baseline
```

### Model $\rightarrow$ Random Forest Classifier

Because their hierarchical structure allows them to learn from both classes, decision trees perform better on imbalanced data than other classification models even though they can be biased towards the dominate class and often have issues of overfitting.

TFIDF was chosen as the vectorizer instead of Count Vectorizer because I want to be able to use the most discriminatory words from each book description in the model instead of just the words that appear often.

There are a lot of hyperparameters to optimize for both TFIDF vectorizer & Random Forest Classifier I decided to feed important hyperparameters into a grid search to get the best possible result.

```python
#set up pipeline
pipe = Pipeline([
    ('tfidf',TfidfVectorizer(stop_words='english')),
    ('rfc',RandomForestClassifier())
])

rfc_params = {
    'tfidf__max_features':[100, 200, 300, 400],
    'tfidf__ngram_range':[(1,1),(1,2),(1,3)],
    'rfc__n_estimators':[150,200,250],
    'rfc__max_depth':[50,75,100,150]
}

rfc_gs = GridSearchCV(pipe, cv = 2, param_grid=rfc_params)
rfc_gs.fit(X_train, y_train)

print("Best Score: ",rfc_gs.best_score_)
print("Best Params: ", rfc_gs.best_params_)
```

```python
print("Train score: ",rfc_gs.score(X_train, y_train))
print(classification_report(y_train, rfc_gs.predict(X_train)))
print("Test score: ", rfc_gs.score(X_test, y_test))
print(classification_report(y_test, rfc_gs.predict(X_test)))
```

### The training score is approximately 1 and is noticably higher than the test score. This probably means the model is overfit on the training data.

```python
conmat = np.array(confusion_matrix(y_test, rfc_gs.predict(X_test)))
confusion = pd.DataFrame(conmat,
                         index = ['is_childrens', 'is_classics','is_fantasy','is_fiction',
                                  'is_historical','is_horror','is_mystery','is_nonfiction',
                                  'is_romance','is_science_fiction','is_young_adult'],
                        columns=['pred_childrens', 'pred_classics','pred_fantasy',
                                 'pred_fiction','pred_historical','pred_horror',
                                 'pred_mystery','pred_nonfiction','pred_romance',
                                 'pred_science_fiction','pred_young_adult'])
confusion
```

```python
plt.subplots(figsize=(10,4))
sns.heatmap(confusion.apply(lambda x: x/x.sum(),axis=1), annot=True, cmap = 'Blues')
None
```

### RFC on full dataset overpredicts fiction which is expected based on the imbalanced data. The model is good at predicting fiction, mystery & fantasy ( > 50% correctly predicted)

```python
# keras model & pipeline
def model_func(input_dim=400, layer_one=200, layer_two=100, layer_three= 50, 
               output_dim=11):
    model = Sequential()
    model.add(Dense(layer_one, input_dim=input_dim, activation='relu'))
    model.add(Dense(layer_two, activation='relu'))
    model.add(Dropout(.5))
    model.add(Dense(layer_three, activation='relu'))
    model.add(Dense(output_dim, activation='sigmoid'))
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model

ff_model = KerasClassifier(build_fn=model_func)

# below class is from Adam Blomberg - converts sparse matrices to dense ones
# https://blomadam.github.io/projects/2017/05/10/apis-lab-starter-code.html
class DenseTransformer(BaseEstimator,TransformerMixin):
    def transform(self, X, y=None, **fit_params):
        return X.todense()
    def fit_transform(self, X, y=None, **fit_params):
        self.fit(X, y, **fit_params)
        return self.transform(X)
    def fit(self, X, y=None, **fit_params):
        return self

pipe = Pipeline([
    ('tfidf',TfidfVectorizer(stop_words='english', max_features=400)),
    ('densify', DenseTransformer()),
    ('model',ff_model)
])

params = {
    'tfidf__ngram_range':[(1,1),(1,2),(1,3)],
    'model__epochs':[10, 15],
    'model__batch_size':[15, 25, 50],
    'model__layer_one':[200],
    'model__layer_two' : [100, 150, 200],
    'model__layer_three':[75, 100, 150]
}
```

```python
y = df['genre_code']
X = df['description']

#split training data into a test & train set
X_train_k, X_test_k, y_train_k, y_test_k = train_test_split(X, y, test_size=0.5, 
                                                            random_state = 87302, 
                                                            stratify=y)
```

```python
gs = GridSearchCV(pipe, param_grid=params)
gs.fit(X_train_k, y_train_k)
```

```python
print("Best Score: ", gs.best_score_)
print("Best Params: ", gs.best_params_)
```

```python
print("Train score: ",gs.score(X_train_k, y_train_k))
print(classification_report(y_train_k, gs.predict(X_train_k)))
print("Test score: ", gs.score(X_test_k, y_test_k))
print(classification_report(y_test_k, gs.predict(X_test_k)))
```

```python
conmat_k = np.array(confusion_matrix(y_test_k, gs.predict(X_test_k)))
confusion_k = pd.DataFrame(conmat_k,
                         index = ['is_childrens', 'is_classics','is_fantasy','is_fiction',
                                  'is_historical','is_horror','is_mystery','is_nonfiction',
                                  'is_romance','is_science_fiction','is_young_adult'],
                        columns=['pred_childrens', 'pred_classics','pred_fantasy',
                                 'pred_fiction','pred_historical','pred_horror',
                                 'pred_mystery','pred_nonfiction','pred_romance',
                                 'pred_science_fiction','pred_young_adult'])
confusion_k
```

```python
plt.subplots(figsize=(10,4))
sns.heatmap(confusion_k.apply(lambda x: x/x.sum(),axis=1), annot=True, cmap='BuGn')
None
```

### In terms of the percentage of each genre that was correctly predicted, Keras model is better than the Random Forest Classifier but slighly worse than the Resampled RFC. It is overpredicting fiction

## Word2Vec
I looked at the top 20 words for each genre with both the TFIDF Vectorizer & Count Vectorizer. There were a handful of words that were common between genres that I do not want to be misclassified as one another such as _romance_ and _mystery_. Both vectorizer cannot provide context for words, only their frequency of occurence. I want to look at Word2Vec as a preprocessing step to be able to provide context to each word and limit the misclassifications that occur because of commonly occuring words.
